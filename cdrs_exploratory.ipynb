```
dogs = pd.read_csv("/data/dogs.csv")
```

```
dogs.head()
```

```
# We only need a few columns
dogs = dogs[['dogID', 'breed', 'pedigree', 'sex', 'dob', 'dateadopted', 'arrivaldate']]

# Remove anything without well formatted dates
dogs_with_dates = dogs[(dogs['dateadopted'] != "0000-00-00") & (dogs['arrivaldate'] != "0000-00-00") & ((dogs['dob'] != "0000-00-00"))]
```

```
dogs_with_dates['dateadopted'] = pd.to_datetime(dogs_with_dates['dateadopted'])
dogs_with_dates['arrivaldate'] = pd.to_datetime(dogs_with_dates['arrivaldate'])
dogs_with_dates['dob'] = pd.to_datetime(dogs_with_dates['dob'])
```

```
dogs_with_dates['time_boarding'] = dogs_with_dates['dateadopted'] - dogs_with_dates['arrivaldate'] 
```
